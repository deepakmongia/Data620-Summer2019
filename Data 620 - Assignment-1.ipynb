```python
import networkx as net
import matplotlib.pyplot as plt
```

```python
Graph1 = net.Graph()
Graph1.add_nodes_from(["Jane", "Ike", "Heather", "Fernando", "Garth", "Carol", "Diane", "Ed", "Andre", "Beverly"])
```

```python
net.info(Graph1)
```

```python
Graph1.nodes()
```

```python
Graph1.add_edges_from([["Jane", "Ike"], ["Heather", "Ike"], ["Heather", "Fernando"], ["Heather", "Garth"], ["Fernando", "Garth"],
                      ["Fernando", "Carol"], ["Fernando", "Andre"], ["Fernando", "Diane"], ["Garth", "Diane"], ["Garth", "Beverly"],
                      ["Garth", "Ed"], ["Carol", "Diane"], ["Carol", "Andre"], ["Ed", "Diane"], ["Ed", "Beverly"], ["Beverly", "Andre"],
                      ["Andre", "Diane"], ["Beverly", "Diane"]])
```

```python
net.draw(Graph1, with_labels=True)
plt.show()
```
