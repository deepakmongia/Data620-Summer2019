## Loading the Data

The research team that curated the dataset provides a file that when loaded, gives a quick visualization of all edges from the "egonet". We can quickly load this with the below code.

```python
#Load the libraries
import networkx as nx
import matplotlib.pyplot as plt

#Create our graph object and display
F = nx.read_edgelist('facebook_combined.txt',
                     create_using=nx.Graph(),
                     nodetype=int)
plt.axis('off')
nx.draw_networkx(F, with_labels=False, node_size=20) 
```

Other things to load into this single networkx graph would be the node features included in the data package as mentioned earlier. Once we map them, we can load all that data into the graph object as metadata for each node.
